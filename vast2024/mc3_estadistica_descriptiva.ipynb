```python
import pandas as pd
```

```python
import json

import chardet

# First, read the file in binary mode
with open('mc3.json', 'rb') as file:
    raw_data = file.read()

# Detect encoding
encoding = chardet.detect(raw_data)['encoding']

# Now, open the file with the detected encoding
with open('mc3.json', 'r', encoding=encoding) as file:
    data = json.load(file)
```

```python
# data.keys() # dict_keys(['directed', 'multigraph', 'graph', 'nodes', 'links'])
# len(data['nodes']) # 60520
# len(data['links']) # 75817
data['nodes']
```

```python
# Convert list of dictionaries to DataFrame
df_nodes = pd.DataFrame(data['nodes'])
df_links = pd.DataFrame(data['links'])
```
